# Branchenomics

## Introduction

A considerable amount of research has gone into increasing the compute capacity of the blockchain protocol, as well as limiting the amount of space consumed in the process.  One proposal is due to Vitalik Buterin - [the *hypercube* architecture](https://blog.ethereum.org/2014/10/21/scalability-part-2-hypercubes/).

We propose an alternate system, based on short-lived blockchains, which we call *branches* (the main ethereum blockchain can be considered the *trunk*).  This comes just as Eris industries has announced their generalized Ðapp client supporting multiple blockchains, which could not have been more timely.

### Mechanism Description

The mechanism is proof of stake based, inspired by the *Slasher* protocol.  A new blockchain is created, which is endowed with a security deposit by the founders.  It is also given an expiry time, which determines when the branch must be terminated.  If the branch is not terminated by that time, all of the endowment is forfeit.  When a branch is terminated, a predetermined number of *verifier signatures* must be writ along with the founders, based on proof of stake.  Unlike Slasher, the difficulty and number of required signatures is set by the founders on the trunk.  The branch chain can use any protocol, however we propose a Slasher type proof of stake based on the trunk.

It can be modeled as the following mechanism:

1. Founders in the new *branch* each contribute a endowment to a contract on the *trunk*, which is withheld until the end of the branches' lifespan.  The endowment represents a security deposite, as well as funds to be used for transactions and transaction fees on the branch chain.  A trunk verifier *reward* is also set aside.

2. Founders set parameters on *trunk*:
    - Trunk verifier difficulty `D*`
    - Number `N*` of required trunk verifiers
    - Expiry time (msecs since epoch)
    - Identifier of branch chain
    
3. Founders create a branch chain. There are many protocols it can follow, however we propose:
    - Slasher style Proof of Stake based on trunk
    - No issuance, just transaction fees
    - Gas price determined at genesis
    - Proof of Stake validation based on stake in trunk at genesis
    - Proof of Stake difficulty `D` determined at genesis
    - Number `N` of verifiers for each block
    - A *trace* of some kind, based aggregate computation done by the branch

4. The branch chain is run to term, the founders write a contract which distributes funds according to the final results of the branch chain

5. Verifiers sign the final result, by updating a map `signatures` with `signatures[account] = sha3(sha3(trace + account))⟩`

6. If there are `N*` signatures such that `sha3(block.header + account) < account.balance ⨉ D*`, the reveal stage is activated for all of the qualifying accounts.

7. All of the verifiers reveal `trace` ; if it is unanimous then the trunk reward is divied up based on stake.  If it is not unanimous, then the founders must update the branch chain so the `trace` is modified and repeat the process starting at step 4.

## Mechanism Justification

We next go over the justification for elements of the above mechanism.

### Expiry Time

If there is no expiry time, then no founder that loses part of their endowment in the branch has any economic incentive to sign anything.  Provided they did not lose all of their endowment in the branch somehow, then the threat of losing their security deposite will force them to sign.

### Trace

Without the trace, the mechanism suffers from the *Nothing at Stake* problem that is the common criticism of proof of stake. If there is no check on whether a verifier has actually verified the correctness of the branch chain, then they can simply sign anything.  By making use of a *Schelling Coin* game based on the state of the computation of the branch, then verifiers must go and actually bother to verify the branch before signing. 

### Unanimity

Finally, we get to the issue of *unanimity*.  This is the primary security measure that the mechanism employs.

We make the following assumption: there is a high probability of untrustworthy or unreliable verifiers.  If we demand that at least *one* reliable verifier be reliable, then the other signers cannot make a profit unless they tell the truth.

In order to try to model how many signers are really required, we ask the following question:

> *For a given Pareto distribution, assuming we cannot trust the top $N\%$ stakeholders, how many signatures do we need to get $P\%$ reliability?*

This may seem severe, however this is something of a worst case scenario for this mechanism.  The top $25\%$ of stakeholders can easily command over $90\%$ of the wealth.  They can easily have their wealth evenly distributed among many accounts, giving them many opertunities to be signers.  This is line with previous proposals by Vitalik, which have aimed at resistance to $25\%$ attacks.

A Pareto distribution is parameterized by a lower limit $x_\text{m}$ and a *Pareto Index* $\alpha$.  Provided that $x > x_\text{m}$, we have:

$$P(X > x) = \left(\frac{x_\text{m}}{x}\right)^\alpha$$

Thus, if we wish to solve for what value of $x$ corresponds to the top $N\%$, then let $n = N/100$ in:

$$x=\frac{x_\text{m}}{\log_\alpha(n)}$$

In the limit, the odds of selecting one of the top $N\%$ as a signer is, assuming $\alpha > 1$:

$$\begin{eqnarray}Odds &=& \frac{\int_{\frac{x_\text{m}}{\log_\alpha(n)}}^\infty x\frac{\alpha x_\text{m}^{\alpha}}{x^{\alpha+1}}}{\int_{x_\text{m}}^\frac{x_\text{m}}{\log_\alpha(n)} x\frac{\alpha x_\text{m}^{\alpha}}{x^{\alpha+1}}}\\
& = &\frac{\int_{\frac{x_\text{m}}{\log_\alpha(n)}}^\infty \frac{1}{x^{\alpha}}}{\int_{x_\text{m}}^\frac{x_\text{m}}{\log_\alpha(n)} \frac{1}{x^{\alpha}}}\\
& = &\frac{\frac{\left(\frac{x_\text{m}}{\log_\alpha(n)}\right)^{1-\alpha}}{1-\alpha}}{\frac{\left(\frac{x_\text{m}}{\log_\alpha(n)}\right)^{1-\alpha}}{1-\alpha} - \frac{x_\text{m}^{1-\alpha}}{1-\alpha}} \\
& = & \frac{1}{1 - \log_\alpha(n)^{1-\alpha}}
\end{eqnarray}$$

The probability $q$ of picking one of the top $N\%$ is related to the $Odds$ by the equation:

$$ Odds = \frac{q}{1-q}$$

Hence:
$$ q = \frac{1}{2 - \log_\alpha(n)^{1-\alpha}}$$

Finally, assuming `sha3` is unbiased, for $K$ verifiers we have the probability of an attack is: 
$$ q^K = \frac{1}{\left(2 - \log_\alpha(n)^{1-\alpha}\right)^K}$$

If we wish to have $P\%$ reliability, then letting $p = P/100$ gives us:

$$\begin{eqnarray}\hat{K} &=& \log_{1/(2 - \log_\alpha(n)^{1-\alpha})}(p)\\
&=& \frac{\log(p)}{-\log\left(2 - \log_\alpha(n)^{1-\alpha}\right)}\end{eqnarray}$$
